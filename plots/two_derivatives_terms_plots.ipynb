```python
#import numpy, pandas and pyplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load two derivatives terms data
data = pd.read_csv("two_derivatives_terms.txt", sep = '\t')
```

```python
def plot_two_derivatives_term(data,term, xlim, ylim):
    term_dic = {"aF": r'$\ddot{\Phi}$', "bF": r'$\ddot{A}_s - \dot{A}_s^2$', "cF": r'$\dot{\Phi}^2$',
               "dF": r'$\dot{A}_s\dot{\Phi}$', "eF": r'$\dot{\tau}^2$'}
    z = np.array(data["z"])
    ter = np.array(data[term])
    plt.plot(z, ter, label = term_dic[term])
    plt.xlim(xlim)
    plt.xlabel("z")
    plt.ylim(ylim)
    plt.ylabel(term_dic[term])
    plt.legend()
    plt.savefig(term+".pdf")
    plt.show()
    
```

```python
plot_two_derivatives_term(data, "aF", (0, 12), (-100, 10))
```

```python
plot_two_derivatives_term(data, "bF", (0, 12), (-10, 100))
```

```python
plot_two_derivatives_term(data, "cF", (0, 12), (-2.5, 50))
```

```python
plot_two_derivatives_term(data, "dF", (0, 12), (-10, 10))
```

```python
plot_two_derivatives_term(data, "eF", (0, 12), (-2.5, 20))
```
