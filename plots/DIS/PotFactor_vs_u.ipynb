```python
# import numpy, matplotlib and pandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
potentials_data = pd.read_csv("PotProfiles.txt", sep = '\t')
interp_pot_data = pd.read_csv("PotInterpolationProfiles.txt", sep = '\t')
# Load u arrays
u_pot = np.array(potentials_data[potentials_data["u"] < 2.5]["u"])
GluePotFac = np.array(potentials_data[potentials_data["u"] < 2.5]["GluePotFac"])
MesonPotFac = np.array(potentials_data[potentials_data["u"] < 2.5]["MesonPotFac"])
u_interp = np.array(interp_pot_data[interp_pot_data["u"] < 2.5]["u"])
GluePotFac_interp = np.array(interp_pot_data[interp_pot_data["u"] < 2.5]["GluePotFac"])
MesonPotFac_interp = np.array(interp_pot_data[interp_pot_data["u"] < 2.5]["MesonPotFac"])
```

```python
plt.plot(u_pot, GluePotFac, label = "Data")
plt.plot(u_interp, GluePotFac_interp, label = "Interpolation")
plt.ylabel("GluonPotFac")
plt.xlabel("u")
plt.title("Gluon Potential Factor")
plt.legend()
plt.show()
```

```python
plt.plot(u_pot, MesonPotFac, label = "Data")
plt.plot(u_interp, MesonPotFac_interp, label = "Interpolation")
plt.ylabel("MesonPotFac")
plt.xlabel("u")
plt.title("Meson Potential Factor")
plt.legend()
plt.show()
```
