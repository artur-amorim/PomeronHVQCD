```python
# import numpy, matplotlib and pandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load raw data
schrodinger_data = pd.read_csv("SchrodingerPotentials.txt", sep = '\t')
fields_data = pd.read_csv("BackgroundFields.txt", sep = '\t')
potentials_data = pd.read_csv("BackgroundPotentials.txt", sep = '\t')
```

```python
# Get z and u data
z = np.array(schrodinger_data["z"])
u = np.array(schrodinger_data["u"])
# Get background potential data
w = np.array(potentials_data["w"])
k = np.array(potentials_data["k"])
Vf = np.array(potentials_data["Vf"])
exp_Phi_3 = np.exp(np.array(fields_data["Phi"])/3)
potFactor = exp_Phi_3 * Vf * w * w
# We just need to compute G
n = len(k)
q = np.array(fields_data["q"])
dtaudA = np.array(fields_data["dtau/dA"])
G = np.sqrt(np.ones(n) + k * dtaudA * dtaudA/ (q * q))
GpotFactor = G * potFactor
```

```python
plt.plot(z, u)
plt.title("u vs z")
plt.xlabel("z")
plt.ylabel("u")
plt.show()
```

```python
plt.plot(z, G)
plt.title("G vs z")
plt.xlabel("z")
plt.ylabel("G")
plt.show()
```

```python
plt.plot(z, GpotFactor, label = r'$G e^{\Phi/3} \, V_f \, w^2$')
plt.plot(z, potFactor, label = r'$e^{\Phi/3} \, V_f \, w^2$')
plt.xlabel("z")
plt.ylabel("pot Factor")
plt.legend()
plt.show()
```

```python
plt.plot(u, potFactor)
plt.title(r'$e^{\Phi/3} \, V_f \, w^2$'+" vs u")
plt.xlabel("u")
plt.ylabel(r'$e^{\Phi/3} \, V_f \, w^2$')
plt.show()
```
