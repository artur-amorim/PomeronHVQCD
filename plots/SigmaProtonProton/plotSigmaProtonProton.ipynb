```python
# import numpy, pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SigmaProtonProtonKernelFixedN400 = pd.read_csv("sigma_proton_proton_kernel_fixed_N_400.txt", sep = '\t')
SigmaProtonProtonKernelFixedN800 = pd.read_csv("sigma_proton_proton_kernel_fixed_N_800_pred.txt", sep = '\t')
SigmaProtonProtonKernelFixedN1000 = pd.read_csv("sigma_proton_proton_kernel_fixed_N_1000_pred.txt", sep = '\t')
SigmaProtonProtonAllSigmaKernelFixedN400 = pd.read_csv("sigma_proton_proton_all_sigma_fixed_kernel_N_400_pred.txt", sep = '\t')
SigmaProtonProtonAllSigmaKernelFixedN800 = pd.read_csv("sigma_proton_proton_all_sigma_fixed_kernel_N_800_pred.txt", sep = '\t')
SigmaProtonProtonAllSigmaKernelFixedN1000 = pd.read_csv("sigma_proton_proton_all_sigma_fixed_kernel_N_1000_pred.txt", sep = '\t')
SigmaProtonProtonAllSigmaN400 = pd.read_csv("sigma_proton_proton_all_sigma_N_400_pred.txt",sep = '\t')
sigmaProtonProton_Es_KernelFixedN1000 = pd.read_csv("sigma_pp_pred_Es.txt", sep = '\t')
exp_data = pd.read_csv("SigmaProtonProton_data_W_lt_10000_without_outliers.txt", sep = '\t')
```

```python
def plotExpDataVsPred(exp_data, pred_data, label, sigma_min, sigma_max, output_file):
    """
        Makes a plot where the experimental points can be compared
        with the predictions of the best fit to those points.
        exp_data : Pandas data frame with the experimental points
        pred_data : Pandas data frame with the best fit predicted points
        label : information about the Pomeron kernel used in the fit
        output_file : name of the file of the saved plot
    """
    plt.errorbar(exp_data["W(GeV)"], exp_data["sigma(mb)"], 
                 xerr = [exp_data["WErrorm"], exp_data["WErrorp"]], 
                 yerr = [exp_data["Errorm"], exp_data["Errorp"]], fmt = "o", label = "exp. data")
    plt.plot(pred_data["W"], pred_data["Pred"], label = label)
    plt.ylabel(r'$\sigma(mb)$')
    plt.xlabel(r'$\sqrt{s}(GeV)$')
    plt.xscale("log")
    plt.title(r'$\sigma$'+" vs " + r'$\sqrt{s}$')
    plt.legend()
    plt.ylim(sigma_min, sigma_max)
    plt.savefig(output_file)
    plt.show()
```

```python
# Plot predictions against experimental data
plotExpDataVsPred(exp_data, SigmaProtonProtonKernelFixedN400, "HVQCD", 0, 200, "SigmaProtonProton_Kernel_Fixed_N_400.pdf")
plotExpDataVsPred(exp_data, SigmaProtonProtonKernelFixedN800, "HVQCD", 0, 200, "SigmaProtonProton_Kernel_Fixed_N_800.pdf")
plotExpDataVsPred(exp_data, SigmaProtonProtonKernelFixedN1000, "HVQCD", 0, 200, "SigmaProtonProton_Kernel_Fixed_N_1000.pdf")
plotExpDataVsPred(exp_data, SigmaProtonProtonAllSigmaKernelFixedN400, "HVQCD", 0, 200, "SigmaProtonProton_All_Sigma_Kernel_Fixed_N_400.pdf")
plotExpDataVsPred(exp_data, SigmaProtonProtonAllSigmaKernelFixedN800, "HVQCD", 0, 200, "SigmaProtonProton_All_Sigma_Kernel_Fixed_N_800.pdf")
plotExpDataVsPred(exp_data, SigmaProtonProtonAllSigmaKernelFixedN1000, "HVQCD", 0, 200, "SigmaProtonProton_All_Sigma_Kernel_Fixed_N_1000.pdf")
plotExpDataVsPred(exp_data, SigmaProtonProtonKernelFixedN400, "HVQCD", 0, 200, "SigmaProtonProton_Kernel_Fixed_N_400.pdf")
plotExpDataVsPred(exp_data, SigmaProtonProtonKernelFixedN800, "HVQCD", 0, 200, "SigmaProtonProton_Kernel_Fixed_N_800.pdf")
plotExpDataVsPred(exp_data, SigmaProtonProtonAllSigmaN400, "HVQCD", 0, 200, "SigmaProtonProton_All_Sigma_N_400.pdf")
```
