```python
# Import numpy, pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wf_data = pd.read_csv("gluon_wf_0.txt",sep = '\t')
plt.plot(wf_data["u"],wf_data["psi"])
plt.ylabel(r'$\psi_0$')
plt.xlabel("u")
plt.title("Gluon "+r'$\psi_0$')
plt.show()
```
