```python
#import numpy
import numpy as np
# import pyplot
import matplotlib.pyplot as plt
# import linear model from scikit learn
from sklearn import linear_model
```

```python
def studyReggeTrajectory(masses = [], label = ""):
    """
        This function computes the Regge trajectory m_n^2 vs n
        given the numerical values in masses.
        Then it plots the trajectory and the experimental values
        and prints the values of the intercept and slope
        of the Regge trajectory.
        The values of the intercept and slope are returned
        masses - is a np.array with the numerical values of m_n
        label - is a string to make the legends of the plots
    """
    # Check we have both numerical values and the label
    n = len(masses)
    if n == 0 or label == "" :
        raise ValueError("Empty list of masses or empty label found in studyReggeTrajectory")
    
    # Find the best linear fit to m_n^2 vs n
    n_vals = np.array(range(1, n + 1)).reshape(-1, 1)
    masses_squared = masses * masses
    reg = linear_model.LinearRegression()
    reg.fit(n_vals, masses_squared)
    
    # Print the intercept and slope variables
    intercept = reg.intercept_
    slope = reg.coef_[0]
    print("Intercept: ", round(intercept,6), " slope: ", round(slope,6))
    
    # Create array with the numerical values of the Regge trajectory
    n_vals = np.linspace(1,n,100)
    regge_trajectory = reg.predict(n_vals.reshape(-1, 1))
    
    # Plot m_n^2 vs n and Regge trajectory
    plt.plot(range(1,n + 1), masses_squared, "bo", label = label + " exp. values")
    plt.plot(n_vals, regge_trajectory, "k", label = label + " trajectory")
    plt.xlabel("n")
    plt.ylabel(r'$m_n^2$')
    plt.legend()
    plt.show()
    return intercept, slope
```

```python
# Define the Regge trajectories from the experimental data.
# These are the masses present in Kiritsis paper in GeV
mrhos = np.array([775.5, 1465, 1720, 1909, 2149, 2265]) / 1000.0     # Non-Singlet Vector Mesons
ma1s = np.array([1230, 1647, 1930, 2096, 2270]) / 1000.0             # Non-Singlet Axial Vector Mesons
mpis = np.array([135, 1300, 1812, 2070, 2360]) / 1000.0              # Non-Singlet Pseudoscalar Mesons
ma0s = np.array([1474, 2025]) / 1000.0                               # Non-Singlet Scalar Mesons
mTG = np.array([2150]) / 1000.0                                      # Singlet Tensor glueballs
momegas = np.array([782.65, 1420, 1670]) / 1000.0                    # Singlet Vector Mesons
mf1s = np.array([1281.9, 1426.4]) / 1000.0                           # Singlet Axial Vector Mesons
```

```python
rho_int, rho_slope = studyReggeTrajectory(mrhos, r'$\rho$')
print(r'$\rho$', " mass squared: ", rho_int + rho_slope, " vs exp. value ", mrhos[0] ** 2)
```

```python
omega_int, omega_slope = studyReggeTrajectory(momegas, r'$\omega$')
print(r'$\omega$', " mass squared: ", omega_int + omega_slope, " vs exp. value ", momegas[0] ** 2)
```
