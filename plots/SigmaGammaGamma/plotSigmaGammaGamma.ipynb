```python
# import numpy, pyplot and pandas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load predicted data sets
sigmaGammaGammaKernelFixedN400 = pd.read_csv("sigma_gamma_gamma_kernel_fixed_N_400_pred.txt", sep = '\t')
sigmaGammaGammaKernelFixedN800 = pd.read_csv("sigma_gamma_gamma_kernel_fixed_N_800_pred.txt", sep = '\t')
sigmaGammaGammaKernelFixedN1000 = pd.read_csv("sigma_gamma_gamma_kernel_fixed_N_1000_pred.txt", sep = '\t')
sigmaGammaGammaAllSigmaKernelFixedN400 = pd.read_csv("sigma_gamma_gamma_all_sigma_kernel_fixed_N_400_pred.txt", sep = '\t')
sigmaGammaGammaAllSigmaKernelFixedN800 = pd.read_csv("sigma_gamma_gamma_all_sigma_fixed_kernel_N_800_pred.txt", sep = '\t')
sigmaGammaGammaAllSigmaKernelFixedN1000 = pd.read_csv("sigma_gamma_gamma_all_sigma_fixed_kernel_N_1000_pred.txt", sep = '\t')
sigmaGammaGammaAllSigmaN400 = pd.read_csv("sigma_gamma_gamma_all_sigma_N_400_pred.txt", sep = '\t')
sigmaGammaGamma_Es_KernelFixedN1000 = pd.read_csv("sigma_gg_pred_Es.txt", sep = '\t')
# Load exp data
exp_data_PDG_W_gt_4 = pd.read_csv("SigmaGammaGamma_PDG_data_W_gt_4.txt", sep = '\t')
```

```python
def plotExpDataVsPred(exp_data, pred_data, label, output_file):
    """
        Makes a plot where the experimental points can be compared
        with the predictions of the best fit to those points.
        exp_data : Pandas data frame with the experimental points
        pred_data : Pandas data frame with the best fit predicted points
        label : information about the Pomeron kernel used in the fit
        output_file : name of the file of the saved plot
    """
    plt.errorbar(exp_data["W(GeV)"], exp_data["sigma(nb)"], 
                 xerr = [exp_data["WErrorm"], exp_data["WErrorp"]], 
                 yerr = [exp_data["Errorm"], exp_data["Errorp"]], fmt = "o", label = "exp. data")
    plt.plot(pred_data["W"], pred_data["Pred"], label = label)
    plt.ylabel(r'$\sigma(nb)$')
    plt.xlabel(r'$\sqrt{s}(GeV)$')
    plt.xscale("log")
    plt.title(r'$\sigma$'+" vs " + r'$\sqrt{s}$')
    plt.legend()
    plt.ylim(0, 1500)
    plt.savefig(output_file)
    plt.show()
```

```python
# Plots predictions vs experimental data
plotExpDataVsPred(exp_data_PDG_W_gt_4, sigmaGammaGammaKernelFixedN400, "HVQCD", "SigmaGammaGamma_Kernel_Fixed_N_400.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_4, sigmaGammaGammaKernelFixedN800, "HVQCD", "SigmaGammaGamma_Kernel_Fixed_N_800.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_4, sigmaGammaGammaKernelFixedN1000, "HVQCD", "SigmaGammaGamma_Kernel_Fixed_N_1000.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_4, sigmaGammaGammaAllSigmaKernelFixedN400, "HVQCD", "SigmaGammaGamma_All_Sigma_Kernel_Fixed_N_400.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_4, sigmaGammaGammaAllSigmaKernelFixedN800, "HVQCD", "SigmaGammaGamma_All_Sigma_Kernel_Fixed_N_800.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_4, sigmaGammaGammaAllSigmaKernelFixedN1000, "HVQCD", "SigmaGammaGamma_All_Sigma_Kernel_Fixed_N_1000.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_4, sigmaGammaGammaAllSigmaN400, "HVQCD", "SigmaGammaGamma_All_Sigma_N_400.pdf")
```
