```python
# import numpy, pyplot and pandas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load datasets with predicted data
SigmaGammaProtonKernelFixedN400 = pd.read_csv("sigma_gamma_proton_kernel_fixed_N_400_pred.txt", sep = '\t')
SigmaGammaProtonKernelFixedN800 = pd.read_csv("sigma_gamma_proton_kernel_fixed_N_800_pred.txt", sep = '\t')
SigmaGammaProtonKernelFixedN1000 = pd.read_csv("sigma_gamma_proton_kernel_fixed_N_1000_pred.txt", sep = '\t')
SigmaGammaProtonAllSigmaKernelFixedN400 = pd.read_csv("sigma_gamma_proton_all_sigma_fixed_kernel_N_400_pred.txt", sep = '\t')
SigmaGammaProtonAllSigmaKernelFixedN800 = pd.read_csv("sigma_gamma_proton_all_sigma_fixed_kernel_N_800_pred.txt", sep = '\t')
SigmaGammaProtonAllSigmaKernelFixedN1000 = pd.read_csv("sigma_gamma_proton_all_sigma_fixed_kernel_N_1000_pred.txt", sep = '\t')
SigmaGammaProtonAllSigmaN400 = pd.read_csv("sigma_gamma_proton_all_sigma_N_400_pred.txt", sep = '\t')
sigmaGammaProton_Es_KernelFixedN1000 = pd.read_csv("sigma_gp_pred_Es.txt", sep = '\t')
# Load exp data
exp_data_PDG_W_gt_461 = pd.read_csv("SigmaGammaP_PDG_data_W_gt_461.txt", sep = '\t')
```

```python
def plotExpDataVsPred(exp_data, pred_data, label, output_file):
    """
        Makes a plot where the experimental points can be compared
        with the predictions of the best fit to those points.
        exp_data : Pandas data frame with the experimental points
        pred_data : Pandas data frame with the best fit predicted points
        label : information about the Pomeron kernel used in the fit
        output_file : name of the file of the saved plot
    """
    plt.errorbar(exp_data["W(GeV)"], exp_data["sigma(mub)"], 
                 xerr = [exp_data["WErrorm"], exp_data["WErrorp"]], 
                 yerr = [exp_data["Errorm"], exp_data["Errorp"]], fmt = "o", label = "exp. data")
    plt.plot(pred_data["W"], pred_data["Pred"], label = label)
    plt.ylabel(r'$\sigma(\mu b)$')
    plt.xlabel(r'$\sqrt{s}(GeV)$')
    plt.xscale("log")
    plt.title(r'$\sigma$'+" vs " + r'$\sqrt{s}$')
    plt.legend()
    plt.ylim(0, 300)
    plt.savefig(output_file)
    plt.show()
```

```python
# Plot the data
plotExpDataVsPred(exp_data_PDG_W_gt_461, SigmaGammaProtonKernelFixedN400, "HVQCD", "SigmaGammaProton_Kernel_Fixed_N_400.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_461, SigmaGammaProtonKernelFixedN800, "HVQCD", "SigmaGammaProton_Kernel_Fixed_N_800.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_461, SigmaGammaProtonKernelFixedN1000, "HVQCD", "SigmaGammaProton_Kernel_Fixed_N_1000.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_461, SigmaGammaProtonAllSigmaKernelFixedN400, "HVQCD", "SigmaGammaProton_All_Sigma_Kernel_Fixed_N_400.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_461, SigmaGammaProtonAllSigmaKernelFixedN800, "HVQCD", "SigmaGammaProton_All_Sigma_Kernel_Fixed_N_800.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_461, SigmaGammaProtonAllSigmaKernelFixedN1000, "HVQCD", "SigmaGammaProton_All_Sigm_Kernel_Fixed_N_1000.pdf")
plotExpDataVsPred(exp_data_PDG_W_gt_461, SigmaGammaProtonAllSigmaN400, "HVQCD", "SigmaGammaProton_All_Sigma_N_400.pdf")
```
