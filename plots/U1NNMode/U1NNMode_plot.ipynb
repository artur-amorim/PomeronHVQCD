```python
# import pandas, numpy and pyplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("U1NNMode_Q2_9000.000000.txt", sep = '\t', skiprows = 1)
plt.plot(data["z"], data["fQ"])
plt.xlabel("z")
plt.ylabel("fQ")
plt.xlim(0, 0.1)
plt.show()
plt.plot(data["z"], data["dfQ/dz"])
plt.xlabel("z")
plt.ylabel("dfQ/dz")
plt.xlim(0, 0.1)
plt.show()
plt.plot(data["z"], data["fQ^2 + (dfQ/dz)^2/Q2"])
plt.xlabel("z")
plt.ylabel("factor")
plt.xlim(0, 0.1)
plt.show()
```
